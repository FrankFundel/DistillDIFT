# Exploring Datasets

There are seperate dataset classes and download bash scripts for each evaluation dataset.

```python
# Whether to load the dataset from hdf5 or from the original images
from_hdf5 = True

# Load the dataset config
dataset_config = read_config('../dataset_config.json')

for config in dataset_config:
    # Load the dataset
    dataset = load_dataset(config, from_hdf5)

    # Get the first image pair from each dataset
    image_pair = dataset[0]

    # Display the image pair
    print("Displaying image pair from dataset: {}".format(config['name']))
    display_image_pair(image_pair)
```
